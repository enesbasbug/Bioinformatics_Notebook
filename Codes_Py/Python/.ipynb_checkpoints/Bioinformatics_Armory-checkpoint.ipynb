```python
file = open('../Datasets/BioArmory/rosalind_ini.txt', 'r')
string = file.read()

A, T, G, C = 0, 0, 0, 0

for base in string:
    if base == 'A':
        A += 1
    elif base == 'T':
        T += 1
    elif base == 'G':
        G += 1
    elif base == 'C':
        C += 1
    else:
        break
print("Answer :", A,C,G,T)
```
