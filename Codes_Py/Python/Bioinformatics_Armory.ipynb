```python
file = open('../Datasets/BioArmory/rosalind_ini.txt', 'r')
string = file.read()

A, T, G, C = 0, 0, 0, 0

for base in string:
    if base == 'A':
        A += 1
    elif base == 'T':
        T += 1
    elif base == 'G':
        G += 1
    elif base == 'C':
        C += 1
    else:
        break
print("Answer :", A,C,G,T)
```

```python
from Bio import Entrez
from Bio import SeqIO

IDs = open("../Datasets/BioArmory/rosalind_frmt.txt",'r').read().split(' ')

handle = Entrez.efetch(db="nucleotide", id=IDs, rettype="fasta")
records = SeqIO.to_dict(SeqIO.parse(handle, "fasta"))
handle.close()

dic = {}
for key, record in records.items():
    dic[record.id] = len(record.seq) 
#print(dic)
fasta = min(dic, key=dic.get)

print('>'+records[fasta].description)
print(records[fasta].seq)
```
